```python
import re
import requests


def save_csv(filename:str, content:list):
    encode = "gbk"
    linbk  = "\n"
    seper  = ","
    with open(filename+".csv", "w", encoding=encode) as f:
        for c in content:
            f.write(seper.join(c))
            f.write(linbk)

def save_tsv(filename:str, content:list):
    encode = "gbk"
    linbk  = "\n"
    seper  = "\t"
    with open(filename+".tsv", "w", encoding=encode) as f:
        for c in content:
            f.write(seper.join(c))
            f.write(linbk)

jobs = ["数据", "证券", "统计"]
url  = "https://search.51job.com/list/040000,000000,0000,00,9,99,%s,2,%d.html"
pattern = "(?<=%s)[\s\S]+?(?=%s)"

for job in jobs:
    job_name = []
    com_name = []
    payment  = []
    for i in range(1, 10+1):
        response = requests.get(url%(job,i))
        # print(response.text)
        text = response.content.decode("gbk")
        # job_name
        tmp  = re.findall("<p class=\"t1 t?g?\d?\">[\s\S]+?</p>", text)
        for i in range(len(tmp)):
            t = re.findall(pattern%("title=\"","\" href=\"http"), tmp[i])
            tmp[i] = t[0]
        job_name += tmp
        # company name
        tmp  = re.findall("<span class=\"t2\">[\s\S]+?</span>", text)
        for i in range(len(tmp),0,-1):
            if tmp[i-1] == "<span class=\"t2\">公司名</span>":
                del tmp[i-1]
        for i in range(len(tmp)):
            t = re.findall(pattern%("title=\"","\" href=\"http"), tmp[i])
            tmp[i] = t[0]
        com_name += tmp
        # payment
        tmp = re.findall("(?<=%s)[\S]*?(?=%s)"%("<span class=\"t4\">","</span>"), text)
        for i in range(len(tmp),0,-1):
            if "薪资" in tmp[i-1]:
                del tmp[i-1]
        payment += tmp

    print(job_name)
    print()
    print(com_name)
    print()
    print(payment)
    print("\n"*10)
    
    save_csv(job, zip(job_name,com_name,payment))
```
