```python
import numpy as np
import pandas as pd

s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
d = {'b' : 1, 'a' : 0, 'c' : 2}
dic = {'a' : 0., 'b' : 1., 'c' : 2.}

print(s)
print(s[s > s.median()])
print(np.exp(s))
print("*"*20)
print(s.index)
print(pd.Series(np.random.randn(6)))
print(pd.Series(d))
print(pd.Series(dic))
```
